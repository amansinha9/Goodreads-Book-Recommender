```python
import numpy as np
import pandas as pd
```

```python
books=pd.read_csv('BX-Books.csv', sep=';',error_bad_lines=False, encoding='latin-1')
```

```python
books
```

```python
books=books[['ISBN','Book-Title','Book-Author']]
```

```python
books
```

```python
books.columns=['ISBN','Book Name','Author']
```

```python
books
```

```python
ratings=pd.read_csv('BX-Book-Ratings.csv', sep=';', encoding='latin-1')
ratings
```

```python
ratings.columns=['user_id','ISBN', 'rating']
```

```python
ratings
```

```python
user_min=ratings['user_id'].value_counts()
```

```python
user_min=user_min[user_min>200]
```

```python
ratings=ratings[ratings['user_id'].isin(user_min.index)]
ratings
```

```python
rating_with_bookname=ratings.merge(books, on='ISBN')
rating_with_bookname
```

```python
book_total_ratings=rating_with_bookname.groupby('Book Name')['rating'].agg('count').reset_index().rename(columns={'rating':'num_ratings'})
```

```python
book_total_ratings
```

```python
ultimate_final=rating_with_bookname.merge(book_total_ratings, on='Book Name')
ultimate_final
```

```python
ultimate_final=ultimate_final[ultimate_final['num_ratings']>=50]
ultimate_final
```

```python
ultimate_final=ultimate_final.drop_duplicates(['user_id','Book Name'])
ultimate_final
```

```python
ultimate_pivot=ultimate_final.pivot_table(values='rating', index='Book Name', columns='user_id').fillna(0)
ultimate_pivot
```

```python
from scipy.sparse import csr_matrix
ultimate_sparse=csr_matrix(ultimate_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
nn=NearestNeighbors(metric='cosine',algorithm='brute')
nn.fit(ultimate_sparse)
```

```python
distances,suggestions=nn.kneighbors(ultimate_pivot.iloc[225,:].values.reshape(1,-1),n_neighbors=6)
```

```python
suggestions=suggestions.ravel()
suggestions
```

```python
for i in range(len(suggestions)):
    print(ultimate_pivot.index[suggestions[i]])
```
